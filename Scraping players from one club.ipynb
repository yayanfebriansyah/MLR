```python
name_players=[]
for i in range (len(soup.find_all("td", {"data-title": "Name"}))):
    if soup.find_all("td", {"data-title": "Name"})[i].text not in name_players:
        name_players.append(soup.find_all("td", {"data-title": "Name"})[i].text)
name_players_1 = [0 for i in range (len(name_players))]
for i in range (len(name_players)):
    name_players_1[i] = unidecode(name_players[i])
    if name_players_1[i][-1] == 'L':
        name_players_1[i] = name_players_1[i][:-1:]

df1['Player Name']=name_players_1
cols=df1.columns.tolist()
cols
cols=cols[-1:]+cols[:-1]
df1[cols]
```
